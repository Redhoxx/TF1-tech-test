```python
def list_values_in_column(column):
    flattened_col = [e for sublist in vectors[column] for e in sublist]
    unique_values = list(set(flattened_col))
    return unique_values
    
def num_unique_values(column):
    flattened_col = [e for sublist in vectors[column] for e in sublist]
    num_unique_values = len(set(flattened_col))
    return num_unique_values
    
def values_in_column(column):
    print(list_values_in_column(column))
    print(num_unique_values(column))
```

```python
values_in_column("device_vector")
values_in_column("channels_vector")
values_in_column("program_genre_vector")
```

```python
from datetime import datetime

def convert_to_period(timestamps):
    periods = []
    for timestamp in timestamps:
        time = datetime.strptime(timestamp, "%Y-%m-%d %H:%M:%S.%f")
        hour = time.hour

        if 6 <= hour < 12:
            periods.append(0)
        elif 12 <= hour < 18:
            periods.append(1)
        elif 18 <= hour < 24:
            periods.append(2)
        else:
            periods.append(3)
    return periods
```

```python
df_encoded = pd.DataFrame()

columns_to_take = ["device_vector", "channels_vector", "program_genre_vector"]

encoders = {}

for column in tqdm(columns_to_take, desc='encoding column'):
    flattened_col = [e for sublist in vectors[column] for e in sublist]
    encoder = LabelEncoder()
    encoder.fit(flattened_col)
    encoders[column] = encoder

    encoded_values = [encoder.transform(sublist) for sublist in vectors[column]]
    df_encoded[column] = encoded_values
    
progressbar_ec = tqdm(total=1, desc='encoding column - ' + 'timestamps_vector')
df_encoded['timestamps_vector'] = vectors['timestamps_vector'].apply(convert_to_period)
progressbar_ec.update(1)

progressbar_ec.close()

df_encoded
```

```python
column_terms = {}

column_terms['timestamps_vector'] = ['matin', 'après-midi', 'soir', 'nuit']
for column, encoder in encoders.items():
    terms = encoder.classes_
    column_terms[column] = terms.tolist()

print(column_terms['timestamps_vector'])
print(column_terms['device_vector'])
print(column_terms['channels_vector'])
print(column_terms['program_genre_vector'])
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

user_gender = df_encoded_sepcol['user_gender']
variables = df_encoded_sepcol.drop('user_gender', axis=1)

variables_transposed = variables.transpose()

scaler = StandardScaler()
variables_std = scaler.fit_transform(variables_transposed)

pca = PCA(n_components=2)
components = pca.fit_transform(variables_std)

components_df = pd.DataFrame(data=components, columns=['PC1', 'PC2'])
components_df = components_df.rename(index=dict(enumerate(variables.columns)))

print(components_df)
```

```python
components_stand = pd.DataFrame()

for col in components_df.columns:
    components_stand[col] = (components_df[col] - components_df[col].mean()) / components_df[col].std()
```

```python
components_stand
```

```python
PC1 = components_stand['PC1']
PC2 = components_stand['PC2']
names = components_stand.index

plt.figure(figsize=(10, 6))

plt.scatter(PC1, PC2)

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Nuage de points pour PC1 et PC2')

for i, name in enumerate(names):
    plt.text(PC1[i], PC2[i], name, fontsize=12)

save_fig("pca_pc1_pc2", save_path="images/pca/")
plt.show()
```

```python
import matplotlib.cm as cm

fig, ax = plt.subplots(figsize=(10, 8))

plt.scatter(PC1, PC2, color='blue')

for x, y, label in zip(PC1, PC2, names):
    plt.arrow(0, 0, x, y)

max_value = max(abs(PC1 + PC2))
ax.set_xlim([-max_value * 1.1, max_value * 1.1])
ax.set_ylim([-max_value * 1.1, max_value * 1.1])

num_colors = len(names)
cmap = cm.get_cmap('tab20', num_colors)

for x, y, label, idx in zip(PC1, PC2, names, range(num_colors)):
    color = cmap(idx)
    plt.scatter(x, y, color=color)
legend_elements = []

for idx, name in enumerate(names):
    color = cmap(idx)
    legend_elements.append(plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=8, label=name))
plt.legend(handles=legend_elements)

plt.xlabel('PC1')
plt.ylabel('PC2')

plt.title('Nuage de points avec flèches pour PC1 et PC2')

plt.grid(True)
save_fig("pca_pc1_pc2_arrows", save_path="images/pca/")
plt.show()
```
